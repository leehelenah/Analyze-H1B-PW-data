Insight 4 - which zipcode has the best rent/price ratio

```python
import os
from spark_utils import *
from pyspark.sql import functions as F

spark = create_spark_session()
bucket = 's3a://helenaudacitybucket'
```

```python
load_zillow = spark.read.parquet(os.path.join(bucket, 'processed_data', 'Zillow_price_rent'))
load_zillow.createOrReplaceTempView('zillow')
```

which zipcode has the lowest rent/price ratio

```python
output = spark.sql("""
SELECT State, Metro, CountyName, Zipcode, 2021_07_Rent, 2021_07_Price, (2021_07_Rent/2021_07_Price) AS Rent_Price_Ratio
FROM zillow
ORDER BY Rent_Price_Ratio
LIMIT 10
""")
output.limit(100).toPandas()
```

*Take home message 1*: If you want to invest in a rental property, avoid the area listed above. For instance 94301 -- Palo Alto and 94040 -- Mountain view

which zipcode has the highest rent/price ratio

```python
output = spark.sql("""
SELECT State, Metro, CountyName, Zipcode, 2021_07_Rent, 2021_07_Price, (2021_07_Rent/2021_07_Price) AS Rent_Price_Ratio
FROM zillow
ORDER BY Rent_Price_Ratio DESC
LIMIT 10
""")
output.limit(100).toPandas()
```

*Take home message 2*: 
It seems that 11937 is a luxury and pricey community. I won't be able to afford that. Most area listed have a very low rent-to-price ratio. However, after I googled them, their high crime rates are high and the school ratings are lower than average. I don't think they are good investments.
